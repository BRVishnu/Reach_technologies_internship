```python
def get_html_file(url): 
    headers = {'User-agent': 'Googlebot/2.1 (+http://www.google.com/bot.html)'}
    source=requests.get(url, headers=headers).text
    soup=BeautifulSoup(source,'lxml')
    return soup
```

```python
def scrapping_data(data):
    data=list(data)
    role=data[0]
    company = data[1]
    posted_date=data[-2]
    
    return company,role,posted_date
```

```python
print(get_html_file(url).prettify)
```

```python
df=pd.DataFrame(columns=["Company Name","Job Role","posted_date"])
```

```python
web_url=["https://www.naukri.com/jobs-in-trivandrum-","1"]
```

```python
k=0
for i in range(1,1000):
    web_url[1]=str(i)
    url="".join(web_url)
    soup=get_html_file(url)
    print(url)
    if soup.find("article",class_="jobTuple bgWhite br4 mb-8")==None:
        break
    else:
        for j in soup.find_all("article",class_="jobTuple bgWhite br4 mb-8"):
            df.loc[k]=scrapping_data(j.stripped_strings)
            k+=1 
```

```python
df.tail()
```

```python
df=df[(df["posted_date"] != '30+ Days Ago')]
```

```python
df["posted_date"].unique()
```

```python
def date_change(day_hour):
    if (day_hour =='Just Now') or (day_hour== 'Today'):
        posted_date=datetime.today().date()
        return posted_date
    else:
        day_hour=day_hour.split(" ")

    if day_hour[1]=="Days":
        posted_date=datetime.today().date()-timedelta(days=int(day_hour[0]))
        return posted_date
    else:
        posted_date=datetime.today().date()-timedelta(days=1)
        return posted_date
        
```

```python
df["posted_date"]=df["posted_date"].apply(date_change)
```
