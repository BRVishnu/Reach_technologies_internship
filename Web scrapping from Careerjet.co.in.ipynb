```python
from bs4 import BeautifulSoup
import requests
```

```python
import pandas as pd
import numpy as np
from datetime import date
from datetime import datetime,timedelta
```

```python
url="https://www.careerjet.co.in/search/jobs?s=&l=Thiruvananthapuram"
```

```python
def get_html_file(url): 
    source=requests.get(url).text
    soup=BeautifulSoup(source,'lxml')
    return soup
```

```python
get_html_file(url).prettify
```

```python
get_html_file(url).body.find("div",id="search").find("div",id="search-content").find_all("article")
```

```python
def web_scrapping(article):
    list2=[]
    for i in article.stripped_strings:
        list2.append(i)
    job_role=list2[0]
    company_name=list2[2]
    job_description=list2[-2]
    posted_date=list2[-1]
                 
    return job_role,company_name,job_description,posted_date
```

```python
df2=pd.DataFrame(columns=["Job role","Company Name","Job Decription","posted_date"])
j=0
```

```python
for i in get_html_file(url).find_all("article"):
    df2.loc[j]=web_scrapping(i)
    j+=1
    
```
