```python
def get_html_file(url):    
    source=requests.get(url,verify=False).text
    soup=BeautifulSoup(source,'lxml')
    return soup
```

### scrapping the data

```python
df1=pd.DataFrame(columns=["Job Role","posted_date"])
df2=pd.DataFrame(columns=["Company name"])
```

```python
j=0
#for is to loop through all the list returned by the find_all function.
for i in get_html_file(url).table.find_all("tr",class_="companyList"):
    job_specific_url="https://www.technopark.org/"+i.td.a.get('href')
    df2.loc[j]=company_name(job_specific_url)
    df1.loc[j]=web_scrapping(job_specific_url)
    j+=1
    print(j,end="")
```

```python
df=pd.concat([df1,df2],axis=1)
```
