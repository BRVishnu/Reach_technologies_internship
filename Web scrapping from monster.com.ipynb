```python
from bs4 import BeautifulSoup
import requests
```

```python
import pandas as pd
import numpy as np
from datetime import date
from datetime import datetime,timedelta
```

```python
url=["https://www.monsterindia.com/search/jobs-in-thiruvananthapuram-trivandrum-","10","?searchId=99024b62-7651-4c2c-b679-9e524997f43b"]
url="".join(url)
```

```python
def get_html_file(url): 
    headers = {'User-agent': 'Googlebot/2.1 (+http://www.google.com/bot.html)'}
    source=requests.get(url, headers=headers).text
    soup=BeautifulSoup(source,'lxml')
    return soup
```

```python
get_html_file(url).prettify
```

```python
def scrapping_data(data):
    data=list(data)
    role=data[0]
    company = data[1]
    posted_date=data[data.index("|")-1]
    
    return company,role,posted_date  
```

```python
web_list=["https://www.monsterindia.com/search/jobs-in-thiruvananthapuram-trivandrum-","10","?searchId=99024b62-7651-4c2c-b679-9e524997f43b"]
```

```python
df=pd.DataFrame(columns=["Company Name","Job Role","posted_date"])
```

```python
k=0
for i in range (1,100):
    web_list[1]=str(i)
    url="".join(web_list)
    soup=get_html_file(url)
    print(url)
    if soup.find("div",class_="card-panel apply-panel job-apply-card")==None:
        print("----------------------------------------------------------")
        break
    else:
        for j in soup.find_all("div",class_="card-panel apply-panel job-apply-card"):
            df.loc[k]=scrapping_data(j.stripped_strings)
            k+=1       
```

```python
df.tail()
```

```python
df=df[(df["posted_date"] != 'Posted: a month ago') & (df["posted_date"] != 'Posted: 2 months ago') & (df["posted_date"] != 'Posted: 3 months ago') & (df["posted_date"] != 'Posted: 4 months ago') & (df["posted_date"] != 'Posted: 5 months ago')]
```

```python
df["posted_date"].unique()
```

```python
def date_change(day_hour):
    day_hour=day_hour.split(" ")
    if day_hour[2]=="hours":
        posted_date=datetime.today().date()
        return posted_date
    elif day_hour[2]=="days":
        posted_date=datetime.today().date()-timedelta(days=int(day_hour[1]))
        return posted_date
    else:
        posted_date=datetime.today().date()-timedelta(days=1)
        return posted_date
        
```

```python
df["posted_date"]=df["posted_date"].apply(date_change)
```
