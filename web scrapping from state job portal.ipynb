```python
from bs4 import BeautifulSoup
import requests
```

```python
import pandas as pd
import numpy as np
from datetime import date
from datetime import datetime,timedelta
```

```python
url="http://www.statejobportal.kerala.gov.in/publicSiteJobs/jobSearch?keywords=&location=thiruvananthapuram&jobTypeName="
```

```python
def get_html_file(url):    
    source=requests.get(url).text
    soup=BeautifulSoup(source,'lxml')
    return soup
```

```python
print(get_html_file(url).prettify())
```

```python
content=get_html_file(url).find('div',{"class":'content'})
```

```python
def web_scrapping(content):
    list2=[]
    for i in content.stripped_strings:
        list2.append(i)
    company_name=list2[2]
    role=list2[0]
    salary=list2[12]
    exp_req=list2[10]
    posted_date=datetime.today().date()-timedelta(int(list2[16].split(" ")[0]))
    return company_name,role,salary,exp_req,posted_date
```

```python
web_scrapping(content)
```

```python
df=pd.DataFrame(columns=["Company Name","Job Role","salary","Required Exp","posted_date"])
```

```python
j=0
for i in get_html_file(url).find_all('div',{"class":'content'}):
    df.loc[j]=web_scrapping(i)
    j+=1
```

```python
df["site"]="statejobportal.kerala.gov.in"
```
